```python
curColor = 0
curLabel = 0
curRegion = 'Asia'
color = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w', 'b', 'g', 'r', 'c', 'm', 'y']
labels = ['Asia', 'CAm', 'EAf', 'Eur', 'MAf', 'ME', 'NAm', 'NAf', 'Oc', 'SAm', 'SAf', 'Car', 'WAf']
colorCountries = [None]*len(families_data_sub)
labelsCountries = [None]*len(families_data_sub)
```

```python
for i in range(len(families_data_sub)):
    if(curRegion != families_data_sub[i,7]):
        curColor = curColor + 1
        curRegion = families_data_sub[i,7]
        curLabel = curLabel +1
    colorCountries[i] = color[curColor] 
    labelsCountries[i] = labels[curLabel]
```
