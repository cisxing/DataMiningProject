```python
#load education data
import sys
import numpy as np
education_region_original = np.loadtxt("C:/Users/Sarah Read/Desktop/Data_Mining/Gender_Data/education_data_labels.csv", delimiter = ',', dtype="string")
```

```python
education_region = education_region_original[1:69,:]
```

```python
#remove quotation marks from data
for i in range(len(education_region)):
   for j in range(size(education_region[0:1])):
        education_region[i-1,j-1] = education_region[i-1,j-1].lstrip('"').rstrip('"')
```

```python
curColor = 0
curLabel = 0
curRegion = 'Asia'
color = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w', 'b', 'g', 'r', 'c', 'm', 'y']
labels = ['Asia', 'CAm', 'EAf', 'Eur', 'MAf', 'ME', 'NAm', 'NAf', 'Oc', 'SAm', 'SAf', 'Car', 'WAf']
colorCountries = [None]*67
labelsCountries = [None]*67
```

```python
for i in range(len(education_region)):
    if(curRegion != education_region[i,20]):
        curColor = curColor + 1
        curRegion = education_region[i,20]
        curLabel = curLabel +1
    colorCountries[i] = color[curColor] 
    labelsCountries[i] = labels[curLabel]
```

```python
cluster_1 = education_region[:,(3,4,20)]
```

```python
import matplotlib
import matplotlib.pyplot as plt
X = cluster_1[:,0]
Y = cluster_1[:,1]
plt.scatter(X,Y,alpha=0.5,s=30)
plt.xlabel("X")
plt.ylabel("Y")
```

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=2, random_state=1)
k_means.fit(cluster_1[:,(0,1)])
y_pred = k_means.predict(cluster_1[:,(0,1)])
plt.scatter(cluster_1[:, 0], cluster_1[:, 1], c=y_pred)
```

```python
k_means = KMeans(n_clusters=13, random_state=1)
k_means.fit(cluster_1[:,(0,1)])
y_pred = k_means.predict(cluster_1[:,(0,1)])
plt.scatter(cluster_1[:, 0], cluster_1[:, 1], c=y_pred)
```

```python
X = cluster_1[:,0]
Y = cluster_1[:,1]
plt.scatter(X,Y,alpha=0.5,s=30, c=colorCountries)
plt.xlabel("X")
plt.ylabel("Y")
```
